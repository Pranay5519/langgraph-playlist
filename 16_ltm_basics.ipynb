```python
from langgraph.store.memory import InMemoryStore
```

```python
store = InMemoryStore()
# creating a namespace
namespace = ("user", "u1")
```

**Creating Memories**

```python
# Adding memories
store.put(namespace, "1", {"data": "User likes pizza"})
store.put(namespace, "2", {"data": "User prefers dark mode"})
```

```python
# another namespace
namespace2 = ("user", "u2")

# Add memories
store.put(namespace2, "1", {"data": "User likes pasta"})
store.put(namespace2, "2", {"data": "User prefers grid style navigation"})
```

**retrieving memories**

```python
store.get(namespace , "1")
```

**retrieving all memories**

```python
items = store.search(namespace2)

for item in items:
    print(item.value)
```

```python
from langchain_ollama import OllamaEmbeddings

embedding_model = OllamaEmbeddings(
    model="nomic-embed-text:latest",
)

store = InMemoryStore(index={'embed': embedding_model, 'dims':1536})
```

```python
store.put(namespace, "1", {"data": "User enjoys learning artificial intelligence and machine learning"})
store.put(namespace, "2", {"data": "User likes building real-world projects in Python"})
store.put(namespace, "3", {"data": "User is interested in deep learning and neural networks"})
store.put(namespace, "4", {"data": "User enjoys exploring Generative AI and LLM-based applications"})
store.put(namespace, "5", {"data": "User likes reading self-improvement and tech-related books"})
store.put(namespace, "6", {"data": "User enjoys hands-on coding more than pure theory"})
store.put(namespace, "7", {"data": "User prefers step-by-step learning with practical examples"})
store.put(namespace, "8", {"data": "User is interested in MLOps, FastAPI, and AI system design"})
store.put(namespace, "9", {"data": "User enjoys solving real-world problems using AI"})
store.put(namespace, "10", {"data": "User likes learning through bullet points and concise explanations"})
```

```python
items = store.search(namespace, query="what is user Good in", limit=1)

for item in items:
    print(item.value)
```
