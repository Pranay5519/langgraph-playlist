```python
import os
from langchain_core.messages import HumanMessage
from tubetalk.retriever import create_retriever_from_url
from tubetalk.chatbot import ChatbotService
```

```python
GOOGLE_API_KEY = os.environ.get("GOOGLE_API_KEY", "YOUR_API_KEY_HERE")
YOUTUBE_URL    = input("🎬 Paste YouTube URL: ").strip()
THREAD_ID      = input("🔑 Session / thread ID (any string): ").strip() or "session-1"
```

```python
retriever = create_retriever_from_url(YOUTUBE_URL)
if retriever is None:
    print("❌ Could not build retriever. Exiting.")
    
```

```python
service = ChatbotService(api_key=GOOGLE_API_KEY)
app     = service.build_chatbot(retriever)
```

```python
 config  = {"configurable": {"thread_id": THREAD_ID}}
```

```python
while True:
    user_input = input("You: ").strip()
    if user_input.lower() in ("quit", "exit", "q"):
        break
    if not user_input:
        continue

    result  = app.invoke(
        {"messages": [HumanMessage(content=user_input)]},
        config=config,
    )

    last_msg = result["messages"][-1].content
    print(f"\n🤖 Bot:\n{last_msg}\n")
```
